```python
import torch
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')
```

```python
from torchvision.transforms.transforms import RandomHorizontalFlip
# Load Existing Dataset                                            
train_data = None
val_data = None
import torchvision.datasets as dset
import torchvision.transforms as T
# from torchvision.transforms import functional as TF

transform = T.Compose([T.Resize((224,224)), T.ToTensor(), T.RandomHorizontalFlip(p=0.5), 
                       T.RandomAdjustSharpness(sharpness_factor=2), T.RandomAutocontrast(), 
                       T.ColorJitter(brightness=0.5, hue=0.3), 
                       T.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])
transform_no_train = T.Compose([T.Resize((224,224)), T.ToTensor(), T.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))])

train_data = dset.Flowers102("./train", split = "test", download=True, transform=transform)
val_data = dset.Flowers102("./val", split='val', download=True, transform=transform_no_train)
test_data = dset.Flowers102("./test", split='train', download=True, transform=transform_no_train)
```

```python
# create mini batches
from torch.utils.data import DataLoader
mini_trains = DataLoader(train_data, batch_size=10, shuffle=True)
mini_vals = DataLoader(val_data, batch_size=10, shuffle=True)
mini_trains
```

```python
from collections.abc import Iterable
iterable = iter(mini_trains)
tensors, labels = next(iterable)
tensors, labels                           # true labels 有10笔
```

```python
from torchvision import models
import torch.nn as nn
import torch.optim as optim
resnet = models.resnet152(pretrained=True).cuda()
num_flatten = resnet.fc.in_features
resnet.fc = nn.Linear(num_flatten, 102)
model = resnet.to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-4)
```

```python
def evaluate(model, mini_vals):
  model.eval()
  with torch.no_grad():
    total_acc = 0
    for x, y in mini_vals:
      x = x.to(device)
      y = y.to(device)
      scores = model(x)
      predictions = scores.max(1)[1]
      acc = predictions.eq(y).sum().item()
      total_acc += acc
    print('Acc:', total_acc/len(val_data))
```

```python
# training data
NUM_EPOCHS = 2
print_every = 25
for epoch in range (NUM_EPOCHS):
  for counter, (x, y) in enumerate(mini_trains):
    # Turn on training mode
    model.train()
    # Move data to the corresponding device
    x = x.to(device)
    y = y.to(device)
    # ForwardProp
    scores = model(x)
    # Calculate Loss
    loss_function = nn.CrossEntropyLoss()
    loss = loss_function(scores, y)
    if counter % print_every == 0:
      print('Epoch', epoch, loss.item(), end=' / ')
      evaluate(model, mini_vals)
    # Clear previous gradient
    optimizer.zero_grad()
    # BackProp
    loss.backward()
    # Update Parameters
    optimizer.step()
```
